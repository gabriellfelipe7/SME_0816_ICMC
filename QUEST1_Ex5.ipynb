```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Dado que $\theta_i$ é o diâmetro médio da cavidade produzida pela esfera $i$, na qual $i = 1,2$. Desta forma, testaremos:

$$
H_0: \theta_1 - \theta_2 <= 0 \\
H_1: \theta_1 - \theta_2 > 0,2
$$

Vamos criar base de dados dada:

image.png

```python
data = pd.DataFrame({'Esfera':['Esfera A', 'Esfera A', 'Esfera A', 'Esfera A', 'Esfera A', 'Esfera A', 'Esfera A', 'Esfera A', 'Esfera A', 'Esfera A',
                               'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B', 'Esfera B'],
                     'Diametro':[7.5, 4.6, 5.7, 4.3, 5.8,3.2,6.1,5.6,3.4,6.5,5.2,4.1,4.3, 4.7, 3.2, 4.9, 5.2, 4.4, 5.7, 6.0]})
data
```

Representação dos diâmetros:

```python
sns.histplot(data=data, x="Diametro", kde=True, bins=10, hue='Esfera')
plt.title('Histograma - Diâmetro médio da cavidade')
```

Representação das diferenças:

```python
Dif = pd.DataFrame({'Diferenças': [2.3, 0.5, 1.4, -0.4, 2.6, -1.7,  0.9, 1.2, -2.3, 0.5]})
sns.histplot(data=Dif, x="Diferenças", kde=True, bins=5)
plt.title('Histograma - Diferenças das Esferas')
```
